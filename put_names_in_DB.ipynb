### Indian Names

```python
inFile=csv.reader(open('indian_names.csv','r'),delimiter='\t')
```

```python
genderCounter=collections.Counter()

for line in inFile:
#    print line[0],line[1].split('|')
    for name in line[1].split('|'):

        genderCounter[line[0]]+=1
        
        detectedLang=langid.classify(name)[0]
        
        nameEnglish=textblob.TextBlob(name.decode('utf-8')).translate().string
        
        if line[0]=='m':
            temp=(name,maxVal,0,False,'IN','asia','hi',detectedLang,nameEnglish)
        elif line[0]=='f':
            temp=(name,0,maxVal,False,'IN','asia','hi',detectedLang,nameEnglish)
        elif line[0]=='unisex':
            temp=(name,0,0,True,'IN','asia','hi',detectedLang,nameEnglish)
        else:
            print 'Unknown gender %s' % line[0]
            
        print name
            
        res=cur.execute(cmd,temp)
        conn.commit()
```
